# Something Interesting
It is observed here that the Actual Rainfall of JUN-SEPT in 1932 and 1954 were lesser than that in 1943 and 1944, 
yet the famine occured from 1944 to 1945... Interesting... 

Also, I have selected JUN-SEPT as they are the months of the monsoon in India.

```python
plt.figure(figsize=(10,6))
plt.plot(rain['YEAR'],rain['Actual Rainfall: JUN-SEPT'],)
plt.tight_layout()
plt.xlabel('Years')
plt.ylabel('Actual Rainfall JUN-SEPT')
```
